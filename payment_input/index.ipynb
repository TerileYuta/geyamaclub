```python
import pandas as pd
import math
```

```python
df = pd.read_excel("data.xlsx", sheet_name = 0)
df.fillna(0)
```

```python
colums = df.columns[6:41]
```

```python
output = []
id = 1
id_list = []
balance = 16000
practice_id = 0

colum_name = colums[len(colums) - 1]

for colum_name in (reversed(colums)):
    practice_id += 1

    colum = []

    for i in df[colum_name][1:11]:
        try:
            if(math.isnan(i) == True):
                i = 0
        except:
            pass

        colum.append(i)

    
    if(colum[1] == 0):
        type_name = "臨時"
    else:
        type_name = "通常練習"

    #参加費
    if(colum[0] > 0):
        action = int(colum[0]) * int(colum[1])
        balance += action

        output.append([
            id,
            type_name,
            "参加費",
            balance,
            "+" + str(action),
            practice_id,
            colum_name.strftime("%Y-%m-%d")
        ])
        
        id_list.append(id)
        id += 1
    
    #その他
    if(colum[2] > 0):
        action = int(colum[2])
        balance += action

        output.append([
            id,
            type_name,
            "その他",
            balance,
            "+" + str(action),
            practice_id,
            colum_name.strftime("%Y-%m-%d")
        ])
        
        id_list.append(id)
        id += 1
    
    #体育館代
    if(colum[7] > 0):
        action = int(colum[7])
        balance -= action

        output.append([
            id,
            type_name,
            "体育館代",
            balance,
            "-" + str(action),
            practice_id,
            colum_name.strftime("%Y-%m-%d")
        ])

        id_list.append(id)
        id += 1
    
    #シャトル代
    if(colum[8] > 0):
        action = int(colum[8])
        balance -= action

        output.append([
            id,
            type_name,
            "シャトル代",
            balance,
            "-" + str(action),
            practice_id,
            colum_name.strftime("%Y-%m-%d")
        ])

        id_list.append(id)
        id += 1
    
    #その他
    if(colum[9] > 0):
        action = int(colum[9])
        balance -= action

        output.append([
            id,
            type_name,
            "その他",
            balance,
            "-" + str(action),
            practice_id,
            colum_name.strftime("%Y-%m-%d")
        ])

        id_list.append(id)
        id += 1


print(len(output))
print(len(id_list))
new_df = pd.DataFrame(output)


```

```python
new_df.to_excel("./new_data.xlsx")
```
